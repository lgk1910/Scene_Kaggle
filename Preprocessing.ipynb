```python
import numpy as np
import os
import cv2
import matplotlib.pyplot as plt
```

```python
import pandas as pd

df = pd.read_csv('./Resources/train-scene classification/train.csv')
```

```python
DATAPATH = "D:/Workspace/Real_Project/Scene_Kaggle/Resources/train-scene classification/train/"
```

```python
# Convert Training Images to matrix and Saving them using pickle
```

```python
X = []
for img in df['image_name']:
    img_path = DATAPATH + img
    x = cv2.imread(img_path)
    new_arr = cv2.resize(x,(150,150,))
    X.append(new_arr)
```

```python
X = np.array(X)
```

```python
X.shape
```

```python
y = np.array(df['label'])
# pickle_out = open("y.pickle","wb")
# pickle.dump(y,pickle_out)
# pickle_out.close()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=13)
```

```python
import pickle
pickle_out = open("y_train.pickle","wb")
pickle.dump(y_train,pickle_out)
pickle_out.close()
pickle_out = open("y_test.pickle","wb")
pickle.dump(y_train,pickle_out)
pickle_out.close()
```

```python
pickle_out = open("X_train.pickle","wb")
pickle.dump(X_train,pickle_out)
pickle_out.close()
pickle_out = open("X_test.pickle","wb")
pickle.dump(X_test,pickle_out)
pickle_out.close()
```

```python
pickle_in = open("X_train.pickle","rb")
X_train = pickle.load(pickle_in)
print(len(X_train))
for img in X_train[:5]:
    plt.imshow(img)
    plt.show()
```
